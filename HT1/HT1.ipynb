### Упражнение 1 - Numpy

1. Пусть у двух матриц совпадает количество строк. Объедините их в одну горизонтально (объедините столбцы этих матриц)

```python
n = 2
m1 = 3
m2 = 4

A = np.zeros((n, m1))
B = np.ones((n, m2))

C = np.hstack((A, B))
C
```

2. Пусть дана numpy-матрица и два массива индексов одинаковой длины. Постройте массив элементов Указание: для выполнения этой операции не требуется никаких именных функций.



```python
n = 3
a = np.array([0,1,2])
b = np.flipud(a)
c = np.zeros(n)
M = np.ones([n,n])
for i in range(n):
    c[i] = M[a[i]][b[i]]
print(c)
```

3. Подсчитайте след матрицы и количество ненулевых элементов на ее диагонали

```python
n = 3
M = 3*np.identity(n)
print(np.trace(M))
print(np.where(M!=0, np.identity(n), np.zeros([n,n])).trace())
```

4. Даны два произвольных numpy-массива одинаковой размерности $(n,m)$. Замените нулями те элементы второго массива, для которых на соответствующем месте в первом массиве находятся отрицательные числа

```python
n = 3
m = 4
A = np.random.randint(-10, 10, size=(n, m))
B = np.ones([n, m])
print(np.where(A<0, np.zeros([n,m]), B))
print(A,B)
```

5. Даны две матрицы с одинаковым количеством строк и количеством столбцов $n$ и $m$ соответственно. Постройте матрицу
всевозможных произведений столбцов друг на друга (в смысле скалярного произведения).

### Упражнение 2 -  Matplotlib
1. Постройте таблицу $2 \times 2$ из четырех графиков (см. функцию plt.subplots)

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

```python
x = np.linspace(0.0001, 10, 1000)
x2 = np.linspace(0.0001, 1, 1000)
y1 = np.log(x)
y2 = np.cos(x)
y3 = x**2 + 4 + np.exp(x)
y4 = sp.special.zeta(x2)

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x2, y4)

# plt.tight_layout()
plt.show()
```
